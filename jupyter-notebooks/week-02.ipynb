```python
import numpy as np
import pandas as pd   
import matplotlib.pyplot as plt
```

```python
# This is a setting to make the plot a big enough and
# the grid lines do not distract the entire plot
plt.rcParams.update(plt.rcParamsDefault)
plt.rcParams.update({
  'font.size': 16,
  'grid.alpha': 0.25})
```

## Read the data

```python
# df is the abbreviation of dataframe, an object to hold a data structure
# table-like format 
# see the documentation of pandas.read_csv() for complete description
df = pd.read_csv("../datasets/gini_rasio_kaltim.csv")      
df
```

## Descriptive statistic

Calculate the average of gini ratio with `np.mean()`.  
We will use this value to draw a horizontal line in the plot
of gini ratio

```python
gini_ratio_avg = np.mean(df["gini_ratio"])
gini_ratio_avg
```

## Create a simple plot

```python
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(df["year"], df["gini_ratio"], marker="o", markersize=8, 
        markerfacecolor="w", markeredgewidth=2, linewidth=2)
ax.axhline(y=gini_ratio_avg, color="r", alpha=0.4)

ax.set_title("Gini ratio of East Kalimantan")
ax.grid("on")
ax.set_xlabel("Year")
ax.set_ylabel("Gini ratio")
plt.show(fig)
```
